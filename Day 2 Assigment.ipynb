Questions 1:List and its default functions

```
#Creating the list
lst=[1,2,3,4]
print(lst)
```

```
#function 1
#Append function
#add element in the existing list
lst.append("python")
print(lst)
```

```
#function 2
#insert function
#add element in the existing list as per your index
lst.insert(2,6)
print(lst)
```

```
#function 3
#remove() function
#remove element from list
lst.remove(1)
print(lst)
```

```
#fuction 4
#pop() function
#removes and returns last element from the list or the given index value
lst.pop(3)
print(lst)
```

```
#fuction 5
#count()  function
#to count the specific element
lst.count(2)
```

Questions 2:Dictionary and its default functions.

```
#create dictionary
dict={1:"apple",2:"mango",3:"orange"}
print(dict)
```

```
#function 1
#copy() function
#returns a shallow copy of the dictionary
x=dict.copy()
print(x)
```

```
#function 2
#keys() function
#return value  of keys from dictonary
dict.keys()
```

```
#function 3
#items() function
#return value  of items from dictonary
dict.items()
```

```
#function 4 
#values() function
#return value  of values from dictonary
dict.values()
```

```
#function 5
#clear() function
#removes all items from the dictionary.
dict.clear()
print(dict)
```

Questions 3
Sets and its default functions.

```
#creating the set
sets={1,2,3,4}
print(sets)
```

```
#function 1
#add() function
#add an element
sets.add(5)
print(sets)
```

```
#function 2
#copy() function
#copy of the set
x=sets.copy()
print(x)
```

```
#function 3
#pop() function
#removes an element from the set
sets.pop()
print(sets)
```

```
#function 4
#remove() function
#removes the specified element
sets.remove(3)
print(sets)
```

```
#function 4
#clear() function
#removes all the elements from the set
sets.clear()
print(sets)
```

Questions 5
Strings and explore default methods

```
#creating string
string="I Love python programing"
print(string)
```

```
#function 1
#capitalize() function
#string with the first character capitalized and rest of the characters in lower case.
string.capitalize()
```

```
#function 1
#capitalize() function
#string with the first character capitalized and rest of the characters in lower case.
string.capitalize()
```

```
#function 2
# lower() function
#converts all the characters of the string to lowercase
print(string.lower())
```

```
#function 3
#upper() function
#converts all the characters of the string to uppercase
print(string.upper())
```

```
#function 4
#title() function
#start with uppercase and the rest of the characters in words are in lowercase.
print(string.title())
```

```
#function 5
#swapcase() function
#lowercase characters got converted to uppercase and vice-versa.
print(string.swapcase())
```
